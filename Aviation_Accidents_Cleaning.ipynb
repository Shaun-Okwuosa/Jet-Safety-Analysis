**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
df.columns
```

The total number of passengers in a flight can be calcuated by adding Total.Serious.Injuries, Total.Fatal.Injuries, Total.Minor.Injuries and Total.Uninjured and the value can be stored in a new column called Total.Passengers

The actions that should be taken before the new column is created are:
* Ensuring all of these columns contain data of a numeric data type. Since the columns contain descrete data, the values should be integers and not floats.
* Dealing with missing values in the columns. Any missing values can be dealt with by filling them in with 0.
* Ensuring that none of these columns contain negative values.



```python
#Filling the missing values with 0.

df[["Total.Serious.Injuries", "Total.Fatal.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]] = (
    df[["Total.Serious.Injuries", "Total.Fatal.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]].fillna(0)
)

# Since all 4 of those columns are floats, they will be converted to int
df[["Total.Serious.Injuries", "Total.Fatal.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]] = (
    df[["Total.Serious.Injuries", "Total.Fatal.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]].astype("int64")
)

#Checking for negative values.
(df[["Total.Serious.Injuries", "Total.Fatal.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]] <0).sum()
```

```python
df["Total.Passengers"] = (
    df["Total.Serious.Injuries"] + df["Total.Fatal.Injuries"] + df["Total.Minor.Injuries"] + df["Total.Uninjured"]
)

# Checking if any values in Total.Passengers is 0
(df["Total.Passengers"] == 0).sum()
```

There are 1,309 rows where the Total.Passengers is 0.

Since this makes up 1.47% of all the data, these rows can be dropped. Removing these rows will improve the quality of injury rate calculations since the accidents will actually involve passengers.

```python
df = df[~(df["Total.Passengers"] == 0)]
```

In order to find out if model is unique to the make, a temporary dataframe called temp will be created. 

Temp will have 3 rows: Make, Model and count.

```python
temp = pd.DataFrame(df.groupby("Make")[["Make", "Model"]].value_counts()).reset_index()
temp["Model"].duplicated().sum()
```
